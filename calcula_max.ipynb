```python
from math import comb
import matplotlib.pyplot as plt
import numpy as np

def count_k_body(n, k, L=3):
    return L*comb(n, k)

def number_of_measurements(n, k, L):
    return L*comb(n, k)

# Exemplo
n = 3
k = 1
print(count_k_body(n, k))
```

```python
count_k_body(2, 3)
```

```python
n = np.arange(2, 10)
n_list = np.arange(2, 10)
k_list = [2, 3, 4, 5 ,6, 10]
```

```python
# Primeiro grfico: linha preta n vs n
plt.plot(n, n, marker='o', color='black', label='n', linewidth=3)

# Outros grficos: para cada valor de k
for k in k_list:
    for L in range(2, 4):
        # y = number_of_measurements(n, k) para cada n  n_list
        yvals = [number_of_measurements(n, k, L) for n in n_list]
        #xvals = [count_k_body(n, k, L) for n in n_list]
        # substituir zeros por NaN para no plotar
        yvals = [np.nan if y == 0 else y for y in yvals]

        plt.plot(n_list, yvals, marker='o', ls='--',label=f'L ={L}, k={k}')

plt.yscale('log')
plt.xscale('log')
plt.xlabel("n (nmero de qubits)")
plt.ylabel("number_of_measurements(n, k)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
